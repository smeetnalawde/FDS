## Statistical Functions

### amin
> This function is used to find the minimum value along a specified axis of an array.
```
numpy.amin(arr, axis=None, out=None, keepdims=<no value>, initial=<no value>, where=<no value>)
```

````python
arr1=np.random.randint(100,size=50).reshape(10,5)

print(arr1)
````

````python
#get minimum of all values
print(np.amin(arr1))
````

````python
#get row of minimum values along each column
print(np.amin(arr1,axis=0))
````

````python
#get column of minimum values along each row

print(np.amin(arr1,axis=1))
````

### amax



````python
print(np.ptp(arr1))
````

````python
#get row of difference between maximum and minimum values along each column
print(np.ptp(arr1,axis=0))
````

````python
#get column of difference between maximum and minimum values along each rows
print(np.ptp(arr1,axis=1))
````

## Trigonometric Functions

### Sine Cosine Tangent
> NumPy provides several trigonometric functions that can be used to calculate the sine, cosine, and tangent of angles in radians.

* np.sin(): This function returns the sine of an angle.

* np.cos(): This function returns the cosine of an angle.

* np.tan(): This function returns the tangent of an angle.

````python
angle=float(input("Enter angle in degrees : "))
rad=angle*(np.pi/180)
````

````python
# sin ()
print(round(np.sin(rad),4))
````

````python
# cos ()
print(round(np.cos(rad),4))
````

````python
# tan ()
print(round(np.tan(rad),4))
````

### arcsin arccos arctan

>NumPy provides three trigonometric functions that compute the inverse of the sine, cosine, and tangent functions, respectively. These functions are:

* ```arcsin(x)``` - Computes the inverse sine (in radians) of the input element-wise.
* ```arccos(x)``` - Computes the inverse cosine (in radians) of the input element-wise.
* ```arctan(x)``` - Computes the inverse tangent (in radians) of the input element-wise.

````python
vals=np.array([0,0.5,1/np.sqrt(2),1])
vals
````

````python
#arcsin
print(np.arcsin(vals))                #gives an array of angles in radians
print(np.arcsin(vals)*(180/np.pi))    #gives an array of angles in degrees
````

````python
#arccos
print(np.arccos(vals))                #gives an array of angles in radians
print(np.arccos(vals)*(180/np.pi))    #gives an array of angles in degrees
````

````python
#arctan
print(np.arctan(vals))                #gives an array of angles in radians
print(np.arctan(vals)*(180/np.pi))    #gives an array of angles in degrees
````

### hypot
>This is a NumPy function that returns the hypotenuse of a right-angled triangle with sides x and y, given by the Pythagorean theorem. It computes the Euclidean distance between the origin and the point (x, y) in a 2D coordinate system. Mathematically, np.hypot(x, y) returns the value of sqrt(x^2 + y^2) for each corresponding element in the input arrays.
```
np.hypot(x, y)
```

````python
#find hypotenuse of triangles with side 3 and 4
print(np.hypot(5,12))
````

````python
#find hypotenuse of triangles with array of side
sides1=np.array([3,5,6,9,15])
sides2=np.array([4,12,8,12,20])
print(np.hypot(sides1,sides2))
````

### Bit Twiddling Functions

>NumPy provides several bit twiddling functions that operate on arrays of binary numbers. These functions include:

* `np.bitwise_and(x1, x2)` - Computes the bitwise AND of two input arrays element-wise.
* `np.bitwise_or(x1, x2)` - Computes the bitwise OR of two input arrays element-wise.
* `np.bitwise_xor(x1, x2)` - Computes the bitwise XOR of two input arrays element-wise.
* `np.invert(x)` - Computes the bitwise NOT of an input array element-wise.
* `np.left_shift(x, shift)` - Shifts the bits of an input array to the left by a specified number of positions.
* `np.right_shift(x, shift)` - Shifts the bits of an input array to the right by a specified number of positions.

>All of these functions take one or two input arrays and return an array of the same shape as the input array(s).

````python
a=np.array([1,2,3,4,5],dtype=np.uint8)
b=np.array([1,2,1,3,6],dtype=np.uint8)
````

````python
#bitwise_and
print(np.bitwise_and(a,b))
````

````python
#bitwise_or
print(np.bitwise_or(a,b))
````

````python
#bitwise_xor
print(np.bitwise_xor(a,b))
````

````python
#invert
print(np.invert(a))
````
