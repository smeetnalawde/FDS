## Plot
```
Axes.plot(*args, scalex=True, scaley=True, data=None, **kwargs)
```
> This is a basic line plot where `x` and `y` are arrays of data that are plotted against each other. It is useful for visualizing trends in data over time or other continuous variables.

````python
year=[2014,2015,2016,2017,2018,2019,2020,2021]
placement_Computer=[80,82,85,90,92,94,96,100]
placement_Civil=[78,80,84,88,90,89,89,88]
````

````python
#plot line graph X axis = year; y- axis= placement statistics. Use the legend for details.

plt.plot(year, placement_Computer, label='Computer')
plt.plot(year, placement_Civil, label='Civil')
plt.xlabel('Year')
plt.ylabel('Placement Statistics')
plt.title('Placement Statistics of Computer and Civil Engineering')
plt.legend()
plt.show()
````

## Barplot
```
Axes.bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)
```

>This is a plot where each data point is represented as a bar. It is useful for comparing data across discrete categories.

> This makes a bar plot.
The bars are positioned at x with the given alignment. Their dimensions are given by height and width. The vertical baseline is bottom (default 0).
Many parameters can take either a single value applying to all bars or a sequence of values, one for each bar.



````python
#plot barplot for same.
bar_width = 0.3
r1 = [i for i in range(len(year))]
r2 = [i + bar_width for i in r1]

plt.bar(r1, placement_Computer, color='green', width=bar_width, edgecolor='white', label='Computer')
plt.bar(r2, placement_Civil, color='red', width=bar_width, edgecolor='white', label='Civil')
plt.xlabel('Year')
plt.ylabel('Placement Statistics')
plt.title('Placement Statistics of Computer and Civil Engineering')
plt.xticks([i + bar_width/2 for i in range(len(year))], year)
plt.legend()
plt.show()
````

## Stem Plot

```
stem([locs,] heads, linefmt=None, markerfmt=None, basefmt=None)
```

> A stem plot draws lines perpendicular to a baseline at each location locs from the baseline to heads, and places a marker there. For vertical stem plots (the default), the locs are x positions, and the heads are y values. For horizontal stem plots, the locs are y positions, and the heads are x values. 

>This is a plot where each data point is represented as a vertical line from the x-axis to the corresponding y-value. It is useful for visualizing changes in data over time or other continuous variables.


````python
#Draw a 3d plot

import numpy as np
import matplotlib.pyplot as plt

z= np.linspace(0, 2*np.pi)
x = np.cos(z-np.pi/2)
y = np.sin(z-np.pi/2)


ax = plt.figure().add_subplot(projection='3d')
ax.stem(x, y, z, label='curve')
ax.legend()
plt.show()
````

## fill_between

```
Axes.fill_between(x, y1, y2=0, where=None, interpolate=False, step=None, *, data=None, **kwargs)
```

> Fill the area between two horizontal curves. The curves are defined by the points (x, y1) and (x, y2). This creates one or multiple polygons describing the filled area. You may exclude some horizontal sections from filling using where. By default, the edges connect the given points directly. Use step if the filling should be a step function, i.e. constant in between x.

````python
x1=np.linspace(0,np.pi)
x2=np.tan(x1)
plt.figure(figsize=(5,3))
plt.fill_between(x1,x2,where=(x2<=0.5), color='red')
plt.fill_between(x1,x2, where=(x2>=0.5),color='green')
plt.show()
````

## Stackplot

```
Axes.stackplot(x, *args, labels=(), colors=None, baseline='zero', data=None, **kwargs)[source]
```

> This is a plot where multiple data sets are stacked on top of each other to show their combined effect. It is useful for visualizing the contribution of each data set to the whole.


````python
days=np.array([1,2,3,4,5,6])
study=np.array([8,9,6,7,7,4])
work=np.array([4,5,4,6,4,5])
sleep=np.array([8,6,7,8,7,9])
extra=np.array([4,4,7,3,6,6])
plt.figure(figsize=(6,4))
plt.stackplot(days,study,work,sleep,extra, labels=['days','study','work','sleep','extra'],colors=['blue','green','yellow','red'])
plt.xlabel('Days')
plt.ylabel('Hours ')
plt.title('Weekly Data')
plt.legend()
plt.show()
````

## Contours

```
Axes.contour(*args, data=None, **kwargs)[source]
```

> This is a plot where the data is represented as a contour plot, which shows the level of a variable at different points on a grid. It is useful for visualizing changes in data over two continuous variables.


````python
x = np.arange(1, 30)
y = x.reshape(-1, 1)
f= x * y**(1/2)

ax = plt.contourf(f, levels=[10, 30, 50],
    colors=['#001a47', '#00235e', '#002b75'], extend='both')
ax.cmap.set_over('#023a9c')
ax.cmap.set_under('#00163b')
ax.changed()
````
