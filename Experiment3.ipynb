## Pandas Data Structures

> Pandas is a popular library for data manipulation and analysis in Python. It provides two primary data structures: Series and DataFrame.

* **Series** 
* **DataFrame**


### Series

>A Series is a one-dimensional labeled array capable of holding any data type. It can be thought of as a column in a spreadsheet or a SQL table. A Series object consists of two parts: the data and the index. The index is a set of labels that identify each element in the data.

Syntax: `pandas.Series(data,index,dtype,copy)`

````python
index=[1,2,3,4,5,6]
data=['a','b','c','d','e','f']

s=pd.Series(data,index=index)
print(s)
````

### DataFrame
>A DataFrame is a two-dimensional labeled data structure with columns of potentially different types. It can be thought of as a spreadsheet or a SQL table. A DataFrame object consists of three parts: the data, the index, and the columns.

Syntax : `pandas.DataFrame( data, index, columns, dtype, copy)`

````python
data = {'name': ['Abhishek', 'Atharv', 'Sayali', 'Vivek'],
        'age': [23, 21, 20, 27],
        'city': ['Patan', 'Kolhapur', 'Sangali', 'Solapur']}
df = pd.DataFrame(data)
print(df)
````

## Slicing
>Slicing in Pandas is similar to slicing in Python, but with some additional functionality. In Pandas, you can slice both rows and columns of a DataFrame using the `df.loc[ ]` and `df.iloc[ ]` methods.

### loc[ ]

> This method is used for label-based slicing. You can slice rows and columns by their labels, rather than their position in the DataFrame. For example, you can use .loc[ ] to select all rows where the index label is equal to a certain value, or to select specific columns by their column labels.

````python
#create a dataframe
data1={'name':['Abhishek','Atharv','Sayali','Prasad'],
       'interest':['photography','football','design','literature'],
       'company':['Sony','INA','Adobe','McGraw Hills']}
#convert above dictionary to dataframe

df1=pd.DataFrame(data1)
print(df1)
````

````python
#for using loc[ ] set index to any key of your choice
df1=df1.set_index('name')
````

````python
print(df1.loc['Abhishek'])
````

### iloc[ ]
> This method is used for position-based slicing. You can slice rows and columns by their position in the DataFrame. For example, you can use .iloc[ ] to select the first five rows, or to select every other column.

````python
# Create lists containing data for each column
emp = [['Akash',25,'Satara'], ['Satish',32,'Karad'], ['Veer',18,'Kolhapur'], ['Dhavan',47,'Pune']]
column=['names','ages','cities']

df2=pd.DataFrame(data=emp,columns=column)
print(df2)
````

````python
#fetching all rows and some columns
print(df2.iloc[:,1:])
````

````python
#fetching some rows and all columns
print(df2.iloc[1:3,:])
````

````python
#fetching all rows and all columns
print(df2.iloc[:,:])
````

````python
#Using Except
print(df2.iloc[:,~df2.columns.isin(['ages'])])
````

````python
#Using Except
print(df2.iloc[:,[0,2]])
````

````python
#Using Except
print(df2.iloc[:,df2.columns!='ages'])
````

## Describe
```
df.describe()
```
>This is a method available on a DataFrame object that provides a statistical summary of the data in the DataFrame. By default, it provides a summary for the numerical columns of the DataFrame, including the count, mean, standard deviation, minimum, and maximum values, as well as the quartile ranges (25th, 50th, and 75th percentiles).


````python
# Create a DataFrame
employee= {
    'name': ['A', 'B', 'C', 'D', 'E'],
    'age': [25, 30, 35, 40, 45],
    'salary': [50000, 60000, 70000, 80000, 90000]
}
df3 = pd.DataFrame(employee)

print(df3)
````

````python
# Call describe() on the DataFrame
print(df3.describe())
````

>In this , we create a simple DataFrame with two columns representing age and salary data. We then call the describe( ) method on the DataFrame object to get a summary of the data. The output provides statistics for each column, including count, mean, standard deviation, minimum, and maximum values, as well as the quartile ranges.
By default, `describe()` only provides summary statistics for numeric data types. If you want to include non-numeric data types in the summary, we can pass the `include='all'` parameter:

````python
print(df3.describe(include='all'))
````
